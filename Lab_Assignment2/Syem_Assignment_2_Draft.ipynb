**Pseudorange** is the *pseudo* distance between a satellite and a satellite signal receiver on the ground. In simple words, it contains the geometric range between the satellite and the receiver, with additional associated errors.

The simplified version of the Pseudorange measurement model is:
$$P=\rho-cdT+ϵ_{sum}$$ and
$$\rho=\sqrt{(x_s - x_r)^2 + (y_s - y_r)^2 + (z_s - z_r)^2}$$
where
$$\rho=geometric\ range$$
$${x_s, y_s, z_s, x_r,y_r,z_r} = 3D\ coordinates\ of\ satellite\ and\ receiver$$
$$cdT = receiver\ clock\ offset$$
$$ϵ_{sum}=general\ errors\ term\ (ϵ_{sum}=\Delta \rho + \Delta_{\text{ion}} + \Delta_{\text{trop}} + \epsilon_P + cdt)$$

\\
Note: the 3D **positions of satellites** are given in this assignment.

### **Unknown States:** Position of receiver & receiver clock offset

We use the psuedorange measurements to estimate the receiver's **3-dimensional position** and the **receiver clock offset:** $$X={x_r,y_r,z_r,cdT}$$.

These will be our states of interest, and we are going to estimate them in the following Least Square Estimation.

```python
# step 1 Import necessary libraries
# Load the psuedorange measurements and satellites' positions files, store them in the format of numpy array
# (hint: use pandas library to parse excel files into dataframe first, then convert them to numpy arrays)
# (1 marks)

import numpy as np
import pandas as pd

# Step 2: Load the pseudorange measurements and satellites' positions files
# Use pandas to read the Excel files into DataFrames
df_pseudorange = pd.read_excel('pseudoranges.xlsx')
df_sat_pos = pd.read_excel('satellite_position.xlsx')

# Step 3: Convert the DataFrames to numpy arrays
pseudorange = df_pseudorange.to_numpy()
sat_pos = df_sat_pos.to_numpy()

# Display the first few entries of each array to verify
print("Pseudorange Measurements (pseudorange):")
print(pseudorange[:5])

print("\nSatellites' Positions (sat_pos):")
print(sat_pos[:5])
```

```python
# step 2
# define the number of measurements n and the number of states m
# (2 mark)

# Define the number of measurements n and the number of states m
n = pseudorange.shape[0]  # Number of pseudo range measurements
m = 4  # Number of states (x, y, z, and receiver clock bias)

# Display the values of n and m to verify
print("Number of measurements (n):", n)
print("Number of states (m):", m)
```

```python
# step 3
# Given that the variance of observation is 1 (m^2), define the weight matrix P
# (1 marks)
P = np.identity(n)

# Display the weight matrix P to verify
print("Weight matrix P:")
print(P)
```

```python
# step 4
# set the initial states as zeros:
# (1 mark)
# Create a zero vector of size m (number of states)
x_0 = np.zeros(m)

# Display the initial states to verify
print("Initial states (x_0):")
print(x_0)
```

```python
# step 5
# GroundTruth coordiantes of the receiver
GroundTruth=np.array([-1633489.37967725,	-3651627.18250363,	4952481.61954918])

# Display the ground truth coordinates to verify
print("Ground Truth coordinates of the receiver (GroundTruth):")
print(GroundTruth)
```

```python
# step 6
# initialize the corrections array as ones, iteration loop counter i as zero and related arrays
# (2 mark)
delta = np.ones(m)  # Corrections array initialized as ones
i = 0  # Iteration loop counter initialized as zero

# Initialize related arrays to store intermediate values
arr_w = []  # Array to store weights
arr_delta = []  # Array to store corrections
arr_err_pos = []  # Array to store position errors
arr_i = []  # Array to store iteration counters

# Copy the initial state
x_hat = x_0.copy()

# Display initialized values to verify
print("Initial Corrections array (delta):")
print(delta)
print("\nInitial Iteration counter (i):")
print(i)
print("\nInitial x_hat:")
print(x_hat)
```

```python
# step 7
# define a function called euclidean_norm, which computes the geometric distances between satellites and receiver
# the input of the function should be two arrays: receiver_pos and satellite_pos
# the function should return an array of distance between satellites and receiver
# you can either use the numpy function or code the math manually
# (1 marks)

def euclidean_norm(receiver_pos, satellite_pos):
    # Compute the Euclidean distances using NumPy
    distance = np.linalg.norm(satellite_pos - receiver_pos, axis=1)
    return distance

# Example usage:
# Assuming satellite_pos is a 2D array where each row is a satellite position
# receiver_pos should be a 1D array with the receiver's coordinates
# distance_array = euclidean_norm(receiver_pos, sat_pos)
```

```python
# step 8
# define a function named design_matrix which computes the design matrix A of this LSE problem
# the input of this funciton should be: estimated states array, satellites positions array and geometric distance array
# the function should return an two-dimensional array which represents the design matrix A
# (2 marks)

def design_matrix(r_pos, s_pos, rho):
    # Number of satellites
    n = s_pos.shape[0]
    
    # Initialize the design matrix A with zeros
    A = np.zeros((n, len(r_pos)))

    # Calculate the design matrix
    for i in range(n):
        # Calculate the partial derivatives
        dx = s_pos[i, 0] - r_pos[0]  # Δx
        dy = s_pos[i, 1] - r_pos[1]  # Δy
        dz = s_pos[i, 2] - r_pos[2]  # Δz
        
        # Compute the norm (distance) from the satellite to the receiver
        norm = np.linalg.norm([dx, dy, dz])
        
        # Fill in the design matrix A
        A[i, 0] = -dx / norm  # ∂ρ/∂x
        A[i, 1] = -dy / norm  # ∂ρ/∂y
        A[i, 2] = -dz / norm  # ∂ρ/∂z
        A[i, 3] = 1           # ∂ρ/∂t (if included in the model)

    return A

# Example usage:
# Assuming you have r_pos, sat_pos, and rho (the geometric distances)
# A_matrix = design_matrix(x_hat, sat_pos, euclidean_norm(x_hat, sat_pos))
```

```python
# step 9
# start the iteration for the LSE (11 marks in total)

# define the convergence criteria for the 'while' loop
# the convergence criteria is that the corrections delta are less than 0.001 (i.e., <0.1 m), for receiver's position x,y,z part
# (1 marks)


while np.linalg.norm(delta[:3]) >= 0.001:  # Check if the norm of delta for x, y, z is greater than or equal to 0.001

    # Compute the geometric distance between Satellites and receiver
    rho = euclidean_norm(x_hat, sat_pos)

    # Compute the design matrix A
    A = design_matrix(x_hat, sat_pos, rho)

    # Compute the misclosure array w
    w = rho - l  # Assuming l contains the observed pseudoranges

    # Compute the corrections delta
    delta = np.linalg.inv(A.T @ A) @ A.T @ w  # Least squares correction

    # Update the states
    x_hat += delta  # Add the corrections to the estimated states

    # Compute the position error using the updated states against ground truth
    err_pos = np.linalg.norm(x_hat[:3] - GroundTruth)  # Compare the x, y, z components

    # Update the counter i
    i += 1

    # Store the misclosure, delta, position error, and counter from each iteration into arrays
    arr_w.append(w)
    arr_delta.append(delta)
    arr_err_pos.append(err_pos)
    arr_i.append(i)



# After convergence, compute the A Posteriori Variance of Unit Weight
# (1 marks)
apv = (w.T @ w) / (len(w) - x_hat.size)  # Adjust based on the number of observations and states

# compute the Covariance matrix of estimated states
# (1 marks)
C_x_hat = apv * np.linalg.inv(A.T @ A)  # Variance-covariance matrix of estimated states
```

```python

# step 10
# Prepare the output data
results = {
    "Estimated States": x_hat,
    "Corrections Array": delta,
    "Estimated Measurements Residuals": w,
    "A Posteriori Variance of Unit Weight": apv,
    "Covariance Matrix of Estimated States": C_x_hat,
    "Number of Iterations": i,
    "3-D Position Errors": err_pos
}

# Print the results to the console
for key, value in results.items():
    print(f"{key}: {value}")
    
# (7 marks)

```

```python
# step 11
# Comment on the computed A Posteriori Variance of Unit Weight value and the Covariance matrix of estimated states
# (2 marks)

print("\n--- Comments on Results ---")
print(f"A Posteriori Variance of Unit Weight (apv): {apv}")
print("This value indicates the estimated variance of the measurement residuals. A lower value suggests a more accurate estimation of the measurements, indicating that the model fits the observations well.")

print(f"\nCovariance Matrix of Estimated States (C_x_hat):\n{C_x_hat}")
print("This matrix provides insights into the uncertainties of the estimated states. The diagonal elements represent the variances of the state estimates, while the off-diagonal elements indicate the covariances between the state estimates. Ideally, smaller values in this matrix suggest higher confidence in the state estimates and less correlation between them.")
```
